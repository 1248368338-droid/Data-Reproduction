```python
# train_decade_models_fixed.py
import os
import re
import jieba
from tqdm import tqdm
from datasets import load_dataset
from gensim.models import Word2Vec

# ===== é…ç½®ï¼šä½¿ç”¨å…¨æ–°ç›®å½• =====
DATA_ROOT = r"D:/äººæ°‘æ—¥æŠ¥å¹´ä»½æ•°æ®"
MODEL_DIR = "models_decade_fixed"  # å…¨æ–°ç›®å½•ï¼Œé¿å…æ··æ·†
DECADE_RANGES = [
    (1950, 1959, "1950s"),
    (1960, 1969, "1960s"),
    (1970, 1979, "1970s"),
    (1980, 1989, "1980s"),
    (1990, 1999, "1990s"),
    (2000, 2009, "2000s"),
    (2010, 2019, "2010s")
]

os.makedirs(MODEL_DIR, exist_ok=True)

# æ·»åŠ å…³é”®æœ¯è¯­ï¼ˆé˜²æ­¢åˆ†è¯é”™è¯¯ï¼‰
SOCIAL_TERMS = [
    "å¦‡å¥³", "å§‘å¨˜", "æ¯äº²", "ç”·äºº", "å¹²éƒ¨", "çˆ¶äº²", "ä¸ˆå¤«",
    "å¤©çœŸ", "æ´»æ³¼", "æ¸©æŸ”", "å¯çˆ±", "æ–‡é™", "åŠ¡å®", "æœæ–­", "å®é™…",
    "æ‹¥å†›ä¼˜å±", "å·¥å†œå…µ", "äººæ°‘å…¬ç¤¾", "æ”¹é©å¼€æ”¾", "å¸‚åœºç»æµ"
]
for term in SOCIAL_TERMS:
    jieba.add_word(term, freq=2000)

def clean_and_tokenize(text):
    text = re.sub(r"[^\u4e00-\u9fa5a-zA-Z0-9ã€‚ï¼Œï¼ï¼Ÿï¼›ï¼š\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    if not text:
        return []
    
    sentences = re.split(r"[ã€‚ï¼ï¼Ÿ\n]", text)
    tokenized = []
    for sent in sentences:
        sent = sent.strip()
        if len(sent) < 5:
            continue
        words = jieba.lcut(sent)
        words = [w for w in words if w.strip() and len(w) >= 1]
        if len(words) >= 3:
            tokenized.append(words)
    return tokenized

# ===== ä¸»è®­ç»ƒå¾ªç¯ =====
for start_year, end_year, decade_name in DECADE_RANGES:
    print(f"\n--- è®­ç»ƒ {decade_name} ({start_year}-{end_year}) ---")
    all_sentences = []
    valid_years = 0

    for year in range(start_year, end_year + 1):
        data_path = os.path.join(DATA_ROOT, f"{year}.jsonl.gz")
        if not os.path.exists(data_path):
            continue
        
        try:
            dataset = load_dataset("json", data_files=data_path, split="train")
            for item in tqdm(dataset, desc=f"  {year}", leave=False):
                all_sentences.extend(clean_and_tokenize(item["text"]))
            valid_years += 1
        except Exception as e:
            print(f"  è·³è¿‡ {year}: {e}")
    
    if not all_sentences:
        print(f"  âš ï¸ æ— æœ‰æ•ˆæ•°æ®ï¼Œè·³è¿‡ {decade_name}")
        continue

    print(f"  åˆå¹¶ {valid_years} å¹´æ•°æ®ï¼Œå…± {len(all_sentences)} å¥å­")

    # ä¿®å¤åçš„å‚æ•°
    model = Word2Vec(
        sentences=all_sentences,
        vector_size=100,      # 100 ç»´è¶³å¤Ÿä¸”é«˜æ•ˆ
        window=8,             # å¢å¤§ä¸Šä¸‹æ–‡çª—å£
        min_count=5,          # å…³é”®ï¼šä¿ç•™ä½é¢‘äººæ ¼è¯
        sg=0,                 # CBOW æ›´é€‚åˆè¯­ä¹‰å…³è”ä»»åŠ¡
        workers=min(4, os.cpu_count()),
        epochs=10,
        seed=42
    )

    model.save(os.path.join(MODEL_DIR, f"{decade_name}.model"))
    print(f"  æ¨¡å‹å·²ä¿å­˜è‡³: {MODEL_DIR}/{decade_name}.model")

print(f"\næ‰€æœ‰ decade æ¨¡å‹è®­ç»ƒå®Œæˆï¼ç›®å½•: {MODEL_DIR}")
```

```python
# train_yearly_models_fixed.py
"""
è‡ªåŠ¨è®­ç»ƒ 1951â€“2019 å¹´ã€Šäººæ°‘æ—¥æŠ¥ã€‹å¹´åº¦è¯å‘é‡ï¼ˆä¿®å¤ç‰ˆï¼‰
- æ¨¡å‹ï¼šCBOW (sg=0) â†’ æ›´é€‚åˆè¯­ä¹‰å…³è”ä»»åŠ¡
- ç»´åº¦ï¼š100ï¼ˆè¶³å¤Ÿä¸”é«˜æ•ˆï¼‰
- çª—å£ï¼š8ï¼ˆæ‰©å¤§ä¸Šä¸‹æ–‡ï¼‰
- min_countï¼š5ï¼ˆä¿ç•™ä½é¢‘äººæ ¼è¯ï¼‰
- è¾“å‡ºï¼šmodels_yearly_fixed/{year}.model
"""

import os
import re
import jieba
from tqdm import tqdm
from datasets import load_dataset
from gensim.models import Word2Vec

# ===== é…ç½® =====
DATA_ROOT = r"D:/äººæ°‘æ—¥æŠ¥å¹´ä»½æ•°æ®"
MODEL_DIR = "models_yearly_fixed"  
YEARS = range(1951, 2020)

os.makedirs(MODEL_DIR, exist_ok=True)

# æ·»åŠ å…³é”®æœ¯è¯­ï¼ˆé˜²æ­¢åˆ†è¯é”™è¯¯ï¼Œå¹¶å¼ºåˆ¶ä¿ç•™äººæ ¼è¯ï¼‰
SOCIAL_AND_TRAIT_TERMS = [
    # æ€§åˆ«ç¾¤ä½“è¯
    "å¦‡å¥³", "å§‘å¨˜", "æ¯äº²", "å¦»å­", "å¥³å„¿",
    "ç”·äºº", "å¹²éƒ¨", "çˆ¶äº²", "ä¸ˆå¤«", "å„¿å­",
    # äººæ ¼ç‰¹è´¨è¯ï¼ˆç¡®ä¿ä¸è¢«åˆ‡åˆ†æˆ–è¿‡æ»¤ï¼‰
    "å¤©çœŸ", "æ´»æ³¼", "æ¸©æŸ”", "å¯çˆ±", "æ–‡é™", "å–„è‰¯", "ä¼˜é›…",
    "åŠ¡å®", "æœæ–­", "å®é™…", "åšå¼º", "ç†æ™º", "å‹‡æ•¢", "ç¨³é‡",
    # ç¤¾ä¼šå†å²æœ¯è¯­
    "æ‹¥å†›ä¼˜å±", "å·¥å†œå…µ", "åœ°ä¸»", "èµ„æœ¬å®¶", "å°‘æ•°æ°‘æ—",
    "åŠ³åŠ¨æ¨¡èŒƒ", "äººæ°‘å…¬ç¤¾", "åˆä½œç¤¾", "åœŸæ”¹", "ç¿»èº«", "é˜¶çº§",
    "å¸å›½ä¸»ä¹‰", "è‹è”", "æ–¯å¤§æ—", "æ–°æ°‘ä¸»ä¸»ä¹‰", "è§£æ”¾åŒº",
    "æ”¹é©å¼€æ”¾", "å¸‚åœºç»æµ", "ä¸‹å²—", "å†œæ°‘å·¥", "ç™½é¢†", "å¯Œäºº",
    "å››åŒ–", "æ€»è·¯çº¿", "å¤§è·ƒè¿›", "æ–‡é©", "çº¢å«å…µ", "çŸ¥é’"
]

for term in SOCIAL_AND_TRAIT_TERMS:
    jieba.add_word(term, freq=2000, tag='n')  # æé«˜è¯é¢‘ï¼Œé˜²æ­¢è¢« min_count è¿‡æ»¤

def clean_and_tokenize(text):
    # ä¿ç•™ä¸­æ–‡ã€å­—æ¯ã€æ•°å­—ã€åŸºæœ¬æ ‡ç‚¹
    text = re.sub(r"[^\u4e00-\u9fa5a-zA-Z0-9ã€‚ï¼Œï¼ï¼Ÿï¼›ï¼š\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    if not text:
        return []
    
    # æŒ‰å¥åˆ†å‰²ï¼ˆå¢å¼ºè¯­ä¹‰å•å…ƒï¼‰
    sentences = re.split(r"[ã€‚ï¼ï¼Ÿ\n]", text)
    tokenized = []
    for sent in sentences:
        sent = sent.strip()
        if len(sent) < 5:  # è¿‡æ»¤å¤ªçŸ­çš„å¥å­
            continue
        words = jieba.lcut(sent)
        words = [w.strip() for w in words if w.strip() and len(w) >= 1]
        if len(words) >= 3:
            tokenized.append(words)
    return tokenized

# ===== ä¸»å¾ªç¯ï¼šé€å¹´è®­ç»ƒ =====
for year in YEARS:
    data_path = os.path.join(DATA_ROOT, f"{year}.jsonl.gz")
    
    if not os.path.exists(data_path):
        print(f"è·³è¿‡ {year}ï¼šæ–‡ä»¶ä¸å­˜åœ¨")
        continue
    
    try:
        print(f"\n--- æ­£åœ¨å¤„ç† {year} å¹´ ---")
        
        # åŠ è½½æ•°æ®
        dataset = load_dataset("json", data_files=data_path, split="train")
        print(f"   åŠ è½½ {len(dataset)} ç¯‡æ–‡ç« ")
        
        # åˆ†è¯
        all_sentences = []
        for item in tqdm(dataset, desc=f"   åˆ†è¯", leave=False):
            all_sentences.extend(clean_and_tokenize(item["text"]))
        
        if len(all_sentences) == 0:
            print(f"   æ— æœ‰æ•ˆå¥å­ï¼Œè·³è¿‡")
            continue
        
        print(f"   ç”Ÿæˆ {len(all_sentences)} ä¸ªå¥å­")
        
        # ä¿®å¤åçš„è®­ç»ƒå‚æ•°
        model = Word2Vec(
            sentences=all_sentences,
            vector_size=100,      # é™ä½ç»´åº¦ï¼Œæå‡è®­ç»ƒæ•ˆç‡å’Œç¨³å®šæ€§
            window=8,             # æ‰©å¤§ä¸Šä¸‹æ–‡çª—å£
            min_count=5,          # å…³é”®ï¼šå…è®¸ä½é¢‘è¯è¿›å…¥ vocab
            sg=0,                 # æ”¹ç”¨ CBOWï¼ˆæ›´é€‚åˆå­¦ä¹ â€œç¾¤ä½“-ç‰¹è´¨â€å…±ç°ï¼‰
            workers=min(6, os.cpu_count()),
            epochs=10,            # å¢åŠ è¿­ä»£æ¬¡æ•°
            seed=42
        )
        
        # ä¿å­˜åˆ°æ–°ç›®å½•
        model.save(os.path.join(MODEL_DIR, f"{year}.model"))
        print(f"   æ¨¡å‹å·²ä¿å­˜è‡³ {MODEL_DIR}/{year}.model")
        
    except Exception as e:
        print(f"   {year} å¹´å‡ºé”™: {e}")
        continue

print(f"\nå…¨éƒ¨å®Œæˆï¼æ–°æ¨¡å‹ä¿å­˜åœ¨ '{MODEL_DIR}' æ–‡ä»¶å¤¹ä¸­ã€‚")
```

```python
# Analysis 1: Gender Valence Asymmetry via AggDiffMAC (Strictly aligned with the paper)
import os
import numpy as np
import pandas as pd
from gensim.models import Word2Vec

# ======================
# é…ç½®ï¼ˆä¸¥æ ¼å¯¹é½è®ºæ–‡ï¼‰
# ======================
MODEL_DIR = "models_decade_sg300w3"  # è¯·ç¡®è®¤ä½ çš„æ¨¡å‹ç›®å½•å
TRAIT_FILE = r"C:/Users/Administrator/Desktop/æ•°æ®å¤ç°æ•°æ®/trait_cn_cleaned.txt"

DECADES = ["1950s", "1960s", "1970s", "1980s", "1990s", "2000s", "2010s"]

# è®ºæ–‡ä¸­çš„æ€§åˆ«æ ‡ç­¾
FEMALE_LABELS = ["å¥³äºº", "å¥³æ€§", "å¥¹", "å¦»å­", "å§‘å¨˜", "å¦‡å¥³"]
MALE_LABELS   = ["ç”·äºº", "ç”·æ€§", "ä»–", "ä¸ˆå¤«", "å°ä¼™å­", "ç”·å­"]

TOP_K_TRAITS = 10  # è®ºæ–‡ Table 1 ä½¿ç”¨ Top 10 traits
MIN_DECADES_APPEAR = 3  # è®ºæ–‡è¦æ±‚ï¼šç‰¹è´¨è¯éœ€åœ¨è‡³å°‘3ä¸ªåå¹´ä¸­å‡ºç°ï¼Œä»¥ä¿è¯æ—¶é—´ç¨³å®šæ€§

# ======================
# åŠ è½½ trait_cn.txt
# ======================
print("åŠ è½½ä¸­æ–‡äººæ ¼ç‰¹è´¨è¯å…¸ (trait_cn.txt)...")
df_traits = pd.read_csv(
    TRAIT_FILE,
    sep=r"\s+",
    header=None,
    names=["Word", "Valence_Mean"],
    encoding="utf-8",
    engine="python"
)

TRAIT_VALENCE = {str(row["Word"]).strip(): float(row["Valence_Mean"]) for _, row in df_traits.iterrows()}
PERSONALITY_TRAITS = set(TRAIT_VALENCE.keys())
print(f"æˆåŠŸåŠ è½½ {len(PERSONALITY_TRAITS)} ä¸ªäººæ ¼ç‰¹è´¨è¯")

# ======================
# å·¥å…·å‡½æ•°
# ======================
def compute_mac(model, group_labels, trait_word):
    """è®¡ç®—ç¾¤ä½“æ ‡ç­¾ä¸ç‰¹è´¨è¯çš„ MAC åˆ†æ•°"""
    sims = []
    for label in group_labels:
        if label in model.wv and trait_word in model.wv:
            sims.append(model.wv.similarity(label, trait_word))
    return np.mean(sims) if sims else None

def compute_diffmac(model, female_labels, male_labels, trait_word):
    """è®¡ç®— DiffMAC = MAC(Female, w) - MAC(Male, w)"""
    mac_f = compute_mac(model, female_labels, trait_word)
    mac_m = compute_mac(model, male_labels, trait_word)
    
    # å¦‚æœä»»ä¸€ MAC ä¸å­˜åœ¨ï¼Œåˆ™ DiffMAC ä¹Ÿä¸å­˜åœ¨
    if mac_f is None or mac_m is None:
        return None
    return mac_f - mac_m

# ======================
# ä¸»åˆ†æï¼šè®¡ç®—æ¯ä¸ªç‰¹è´¨è¯çš„ AggDiffMAC
# ======================
# æ­¥éª¤1: æ”¶é›†æ¯ä¸ªç‰¹è´¨è¯åœ¨æ¯ä¸ªåå¹´çš„ DiffMAC
trait_diffmac_history = {trait: [] for trait in PERSONALITY_TRAITS}

for decade in DECADES:
    print(f"å¤„ç† {decade}...")
    model_path = os.path.join(MODEL_DIR, f"{decade}.model")
    if not os.path.exists(model_path):
        print(f"  è·³è¿‡ {decade}ï¼šæ¨¡å‹ä¸å­˜åœ¨")
        continue
    
    try:
        model = Word2Vec.load(model_path)
    except Exception as e:
        print(f"  è·³è¿‡ {decade}ï¼šæ¨¡å‹åŠ è½½å¤±è´¥ - {e}")
        continue

    female_avail = [w for w in FEMALE_LABELS if w in model.wv]
    male_avail   = [w for w in MALE_LABELS if w in model.wv]
    
    if not female_avail or not male_avail:
        print(f"  è·³è¿‡ {decade}ï¼šç¼ºå°‘æ€§åˆ«æ ‡ç­¾")
        continue

    for trait in PERSONALITY_TRAITS:
        diffmac = compute_diffmac(model, female_avail, male_avail, trait)
        if diffmac is not None:
            trait_diffmac_history[trait].append(diffmac)
        # å¦‚æœä¸º Noneï¼Œå°±ä¸è®°å½•ï¼Œåç»­ä¼šè¿‡æ»¤æ‰

# æ­¥éª¤2: è®¡ç®— AggDiffMAC å¹¶ç­›é€‰
# AggDiffMAC = mean(DiffMAC over time)ï¼Œä½†åªè€ƒè™‘å‡ºç° >= MIN_DECADES_APPEAR çš„è¯
trait_aggdiffmac = {}
for trait, diffmac_list in trait_diffmac_history.items():
    if len(diffmac_list) >= MIN_DECADES_APPEAR:
        trait_aggdiffmac[trait] = np.mean(diffmac_list)

# æ­¥éª¤3: è·å–æœ€å¥³æ€§åŒ–å’Œæœ€ç”·æ€§åŒ–çš„ Top K ç‰¹è´¨
female_top_traits = sorted(trait_aggdiffmac.items(), key=lambda x: x[1], reverse=True)[:TOP_K_TRAITS]
male_top_traits = sorted(trait_aggdiffmac.items(), key=lambda x: x[1])[:TOP_K_TRAITS]

# æ­¥éª¤4: è®¡ç®—å®ƒä»¬çš„å¹³å‡æ•ˆä»·
female_valence = np.mean([TRAIT_VALENCE[trait] for trait, _ in female_top_traits])
male_valence = np.mean([TRAIT_VALENCE[trait] for trait, _ in male_top_traits])

# ======================
# æ‰“å°å¹¶ä¿å­˜ç»“æœ (æ¨¡ä»¿è®ºæ–‡ Table 1)
# ======================
print("\n" + "="*80)
print("å¤ç°è®ºæ–‡ Analysis 1: æ€§åˆ«æ•ˆä»·ä¸å¯¹ç§° (åŸºäº AggDiffMAC)")
print("="*80)
print(f"{'Category':<10} {'Group':<10} {'Valence':<10} {'Top 10 Traits'}")
print("-"*80)

# è¾“å‡ºå¥³æ€§è¡Œ
female_traits_str = ", ".join([trait for trait, _ in female_top_traits])
print(f"{'Gender':<10} {'woman':<10} {female_valence:<10.2f} {female_traits_str}")

# è¾“å‡ºç”·æ€§è¡Œ
male_traits_str = ", ".join([trait for trait, _ in male_top_traits])
print(f"{'Gender':<10} {'man':<10} {male_valence:<10.2f} {male_traits_str}")

# ä¿å­˜è¯¦ç»†ç»“æœåˆ°CSV
results = []
for trait, score in female_top_traits:
    results.append({"Group": "woman", "Trait": trait, "AggDiffMAC": score, "Valence": TRAIT_VALENCE[trait]})
for trait, score in male_top_traits:
    results.append({"Group": "man", "Trait": trait, "AggDiffMAC": score, "Valence": TRAIT_VALENCE[trait]})

df_result = pd.DataFrame(results)
output_file = "analysis1_gender_valence_replication_AggDiffMAC.csv"
df_result.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"\nè¯¦ç»†ç»“æœå·²ä¿å­˜è‡³: {os.path.abspath(output_file)}")
```

```python
# Reimplementation of Analysis 2: Temporal Evolution (Gender Only, Bidirectional)
# Computes valence for both most feminine and most masculine traits per decade

import os
import numpy as np
import pandas as pd
from gensim.models import Word2Vec

# ======================
# é…ç½®
# ======================
MODEL_DIR = "models_decade_fixed"
TRAIT_FILE = r"C:/Users/Administrator/Desktop/æ•°æ®å¤ç°æ•°æ®/trait_cn_cleaned.txt"

DECADES = ["1950s", "1960s", "1970s", "1980s", "1990s", "2000s", "2010s"]

FEMALE_LABELS = ["å¥³äºº", "å¥³æ€§", "å¥¹", "å¦»å­", "å§‘å¨˜", "å¦‡å¥³"]
MALE_LABELS   = ["ç”·äºº", "ç”·æ€§", "ä»–", "ä¸ˆå¤«", "å°ä¼™å­", "ç”·å­"]

TOP_K = 10

# ======================
# åŠ è½½æ¸…æ´—åçš„ç‰¹è´¨è¯è¡¨
# ======================
print("åŠ è½½æ¸…æ´—åçš„ä¸­æ–‡äººæ ¼ç‰¹è´¨è¯å…¸...")
df_traits = pd.read_csv(
    TRAIT_FILE,
    sep=r"\s+",
    header=None,
    names=["Word", "Valence_Mean"],
    encoding="utf-8",
    engine="python"
)

TRAIT_VALENCE = {str(row["Word"]).strip(): float(row["Valence_Mean"]) for _, row in df_traits.iterrows()}
PERSONALITY_TRAITS = set(TRAIT_VALENCE.keys())
print(f"åŠ è½½ {len(PERSONALITY_TRAITS)} ä¸ªäººæ ¼ç‰¹è´¨è¯")

# ======================
# å·¥å…·å‡½æ•°ï¼šè®¡ç®— MAC
# ======================
def compute_mac(model, labels, trait):
    sims = []
    for label in labels:
        if label in model.wv and trait in model.wv:
            sims.append(model.wv.similarity(label, trait))
    return np.mean(sims) if sims else None

# ======================
# ä¸»åˆ†æï¼šæŒ‰ decade è®¡ç®— DiffMAC å¹¶æå– Top 10 å¥³æ€§åŒ– & ç”·æ€§åŒ–ç‰¹è´¨
# ======================
results = []

for decade in DECADES:
    model_path = os.path.join(MODEL_DIR, f"{decade}.model")
    if not os.path.exists(model_path):
        print(f"è·³è¿‡ {decade}ï¼šæ¨¡å‹ç¼ºå¤±")
        continue
    
    try:
        model = Word2Vec.load(model_path)
    except Exception as e:
        print(f"è·³è¿‡ {decade}ï¼šåŠ è½½å¤±è´¥ - {e}")
        continue

    female_avail = [w for w in FEMALE_LABELS if w in model.wv]
    male_avail   = [w for w in MALE_LABELS if w in model.wv]

    if not female_avail or not male_avail:
        print(f"è·³è¿‡ {decade}ï¼šæ€§åˆ«æ ‡ç­¾ä¸è¶³")
        continue

    # è®¡ç®—æ‰€æœ‰ç‰¹è´¨çš„ DiffMAC = MAC(woman) - MAC(man)
    diffmac_scores = []
    for trait in PERSONALITY_TRAITS:
        mac_f = compute_mac(model, female_avail, trait)
        mac_m = compute_mac(model, male_avail, trait)
        if mac_f is not None and mac_m is not None:
            diffmac = mac_f - mac_m
            diffmac_scores.append((trait, diffmac))
    
    if len(diffmac_scores) < TOP_K * 2:
        print(f"è·³è¿‡ {decade}ï¼šæœ‰æ•ˆç‰¹è´¨è¯ä¸è¶³")
        continue

    # æ’åº
    diffmac_scores.sort(key=lambda x: x[1], reverse=True)

    # æœ€å¥³æ€§åŒ–ï¼šDiffMAC æœ€é«˜ï¼ˆTop 10ï¼‰
    top_feminine = [trait for trait, _ in diffmac_scores[:TOP_K]]
    valence_fem = np.mean([TRAIT_VALENCE[t] for t in top_feminine])

    # æœ€ç”·æ€§åŒ–ï¼šDiffMAC æœ€ä½ï¼ˆå³æœ€è´Ÿï¼ŒBottom 10 â†’ å–æœ€å10ä¸ªï¼‰
    top_masculine = [trait for trait, _ in diffmac_scores[-TOP_K:]]
    valence_mas = np.mean([TRAIT_VALENCE[t] for t in top_masculine])

    results.append({
        "Decade": decade,
        "Female_Valence": round(valence_fem, 2),
        "Male_Valence": round(valence_mas, 2),
        "Female_Top_Traits": ", ".join(top_feminine),
        "Male_Top_Traits": ", ".join(top_masculine)
    })

    print(f" {decade} | å¥³: {valence_fem:.2f} | ç”·: {valence_mas:.2f}")

# ======================
# ä¿å­˜ç»“æœ
# ======================
df = pd.DataFrame(results)
output_file = "analysis2_gender_diffmac_bidirectional.csv"
df.to_csv(output_file, index=False, encoding="utf-8-sig")

# ======================
# æ‰“å°è¡¨æ ¼ï¼ˆå¯¹æ ‡è®ºæ–‡ Fig 2b çš„åŒå‘è¶‹åŠ¿ï¼‰
# ======================
print("\n" + "="*90)
print("Analysis 2: Decade-wise Valence of Gender-Stereotypical Traits (by DiffMAC)")
print("="*90)
print(df.to_string(index=False, formatters={
    "Female_Valence": "{:.2f}".format,
    "Male_Valence": "{:.2f}".format
}))

# ======================
# ç‰¹åˆ«è¾“å‡ºå…³é”®å¹´ä»£ï¼ˆå¦‚ Table 2 æ‰€ç¤ºï¼‰
# ======================
print("\n" + "-"*60)
print("é‡ç‚¹å¹´ä»£ï¼š1960sâ€“1980sï¼ˆå¥³æ€§ç‰¹è´¨ï¼Œå¯¹æ ‡è®ºæ–‡ Table 2ï¼‰")
print("-"*60)
for _, row in df.iterrows():
    if row["Decade"] in ["1960s", "1970s", "1980s"]:
        print(f"{row['Decade']} | å¥³æ€§æ•ˆä»·: {row['Female_Valence']}")
        print(f"  å¥³æ€§ Top 10: {row['Female_Top_Traits']}\n")

print(f"\nå®Œæ•´ç»“æœå·²ä¿å­˜è‡³: {os.path.abspath(output_file)}")
```

```python
# Analysis 3: Quantifying social change through year-to-year embeddings é€šè¿‡é€å¹´åµŒå…¥é‡åŒ–ç¤¾ä¼šå˜è¿
TRAIT_FILE = r"C:/Users/Administrator/Desktop/æ•°æ®å¤ç°æ•°æ®/trait_cn_cleaned.txt"

# replicate_analysis3_valence.py
import os
import numpy as np
import pandas as pd
from gensim.models import Word2Vec

# é…ç½®
YEARLY_MODEL_DIR = "models_yearly_fixed"
TRAIT_FILE = r"C:/Users/Administrator/Desktop/æ•°æ®å¤ç°æ•°æ®/trait_cn_cleaned.txt"
YEARS = list(range(1950, 2020))

FEMALE_LABELS = ["å¥³äºº", "å¥³æ€§", "å¥¹", "å¦»å­", "å§‘å¨˜", "å¦‡å¥³"]
MALE_LABELS   = ["ç”·äºº", "ç”·æ€§", "ä»–", "ä¸ˆå¤«", "å°ä¼™å­", "ç”·å­"]

TOP_K = 10  # å– DiffMAC æœ€é«˜çš„å‰10ä¸ªç‰¹è´¨

# åŠ è½½ç‰¹è´¨è¯åŠæ•ˆä»·
df_traits = pd.read_csv(
    TRAIT_FILE,
    sep=r"\s+",
    header=None,
    names=["Word", "Valence"],
    engine="python",
    encoding="utf-8"
)
TRAIT_VALENCE = {str(row["Word"]).strip(): float(row["Valence"]) for _, row in df_traits.iterrows()}
PERSONALITY_TRAITS = list(TRAIT_VALENCE.keys())
print(f"åŠ è½½ {len(PERSONALITY_TRAITS)} ä¸ªæ¸…æ´—åçš„äººæ ¼ç‰¹è´¨è¯")

# å·¥å…·å‡½æ•°ï¼šè®¡ç®— MAC
def compute_mac(model, labels, trait):
    sims = []
    for label in labels:
        if label in model.wv and trait in model.wv:
            sims.append(model.wv.similarity(label, trait))
    return np.mean(sims) if sims else None

# ä¸»å¾ªç¯ï¼šé€å¹´è®¡ç®—å¥³æ€§æ•ˆä»·ï¼ˆåŸºäº DiffMAC = MAC(woman) - MAC(man)ï¼‰
results = []

for year in YEARS:
    model_path = os.path.join(YEARLY_MODEL_DIR, f"{year}.model")
    if not os.path.exists(model_path):
        continue
    try:
        model = Word2Vec.load(model_path)
    except:
        continue

    # è¿‡æ»¤æ¨¡å‹ä¸­å­˜åœ¨çš„æ ‡ç­¾
    female_avail = [w for w in FEMALE_LABELS if w in model.wv]
    male_avail   = [w for w in MALE_LABELS if w in model.wv]
    
    if not female_avail or not male_avail:
        continue

    # è®¡ç®—æ¯ä¸ªç‰¹è´¨çš„ DiffMAC
    diffmac_list = []
    for trait in PERSONALITY_TRAITS:
        mac_f = compute_mac(model, female_avail, trait)
        mac_m = compute_mac(model, male_avail, trait)
        if mac_f is not None and mac_m is not None:
            diffmac_list.append((trait, mac_f - mac_m))
    
    if len(diffmac_list) < TOP_K:
        continue

    # å– DiffMAC æœ€é«˜çš„ TOP_K ä¸ªç‰¹è´¨ï¼ˆæœ€èƒ½ä»£è¡¨â€œå¥³æ€§ vs ç”·æ€§â€çš„æ­£é¢ç‰¹è´¨ï¼‰
    diffmac_list.sort(key=lambda x: x[1], reverse=True)
    top_traits = [t for t, _ in diffmac_list[:TOP_K]]
    
    # è®¡ç®—å¹³å‡æ•ˆä»·
    mean_valence = np.mean([TRAIT_VALENCE[t] for t in top_traits])
    
    results.append({
        "Year": year,
        "Female_Valence": round(mean_valence, 3),
        "Top_Traits": ", ".join(top_traits)
    })

# è½¬ä¸º DataFrame
df = pd.DataFrame(results)
output_csv = "analysis3_female_valence_yearly.csv"
df.to_csv(output_csv, index=False, encoding="utf-8-sig")

# æ‰“å°å…³é”®å¹´ä»£
print("\n=== å…³é”®å¹´ä»£å¥³æ€§æ•ˆä»· ===")
for y in [1960, 1965, 1966, 1970, 1976, 1980, 2000, 2010]:
    row = df[df["Year"] == y]
    if not row.empty:
        print(f"{y}: {row['Female_Valence'].values[0]:.2f} | {row['Top_Traits'].values[0][:50]}...")

print(f"\nâœ… ç»“æœå·²ä¿å­˜è‡³: {os.path.abspath(output_csv)}")
```

```python
#ç»˜åˆ¶çƒ­åŠ›å›¾
import os
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
import seaborn as sns

# ======================
# é…ç½®ï¼ˆä¸ä½ ä¹‹å‰ä¸€è‡´ï¼‰
# ======================
YEARLY_MODEL_DIR = "models_yearly_fixed"   
TRAIT_FILE = r"C:/Users/Administrator/Desktop/æ•°æ®å¤ç°æ•°æ®/trait_cn_cleaned.txt"
YEARS = list(range(1950, 2020))

FEMALE_LABELS = ["å¥³äºº", "å¥³æ€§", "å¥¹", "å¦»å­", "å§‘å¨˜", "å¦‡å¥³"]
MALE_LABELS   = ["ç”·äºº", "ç”·æ€§", "ä»–", "ä¸ˆå¤«", "å°ä¼™å­", "ç”·å­"]

# åŠ è½½ç‰¹è´¨è¯
df_traits = pd.read_csv(TRAIT_FILE, sep=r"\s+", header=None, names=["Word", "Valence"], engine="python")
PERSONALITY_TRAITS = [str(w).strip() for w in df_traits["Word"].tolist()]
BLACKLIST = {"å¥½è‰²", "å¼±æ™º", "çŒ¥ç", "ç—´å‘†"}
PERSONALITY_TRAITS = [w for w in PERSONALITY_TRAITS if w not in BLACKLIST]

# ======================
# å·¥å…·å‡½æ•°
# ======================
def get_association_vector(model, labels, traits):
    vec = []
    for trait in traits:
        sims = []
        for label in labels:
            if label in model.wv and trait in model.wv:
                sims.append(model.wv.similarity(label, trait))
        avg_sim = np.mean(sims) if sims else np.nan
        vec.append(avg_sim)
    return np.array(vec)

# ======================
# é‡æ–°åŠ è½½å‘é‡ï¼ˆç¡®ä¿ä¸çƒ­åŠ›å›¾ä¸€è‡´ï¼‰
# ======================
female_vectors = {}
male_vectors = {}

for year in YEARS:
    model_path = os.path.join(YEARLY_MODEL_DIR, f"{year}.model")
    if not os.path.exists(model_path):
        continue
    try:
        model = Word2Vec.load(model_path)
    except:
        continue

    female_avail = [w for w in FEMALE_LABELS if w in model.wv]
    male_avail   = [w for w in MALE_LABELS if w in model.wv]
    if not female_avail or not male_avail:
        continue

    fv = get_association_vector(model, female_avail, PERSONALITY_TRAITS)
    mv = get_association_vector(model, male_avail, PERSONALITY_TRAITS)

    if np.all(np.isnan(fv)) or np.all(np.isnan(mv)):
        continue

    female_vectors[year] = fv
    male_vectors[year] = mv

# ======================
# æ„å»ºå¹´ä»½åˆ—è¡¨å’Œç›¸å…³çŸ©é˜µ
# ======================
years_list = sorted(set(female_vectors.keys()) & set(male_vectors.keys()))
n = len(years_list)

# åˆå§‹åŒ–ç›¸å…³çŸ©é˜µ
corr_mat_f = np.full((n, n), np.nan)
corr_mat_m = np.full((n, n), np.nan)

# è®¡ç®— Pearson ç›¸å…³
for i, y1 in enumerate(years_list):
    for j, y2 in enumerate(years_list):
        # å¥³æ€§
        v1, v2 = female_vectors[y1], female_vectors[y2]
        valid = ~(np.isnan(v1) | np.isnan(v2))
        if np.sum(valid) >= 10:
            r, _ = pearsonr(v1[valid], v2[valid])
            corr_mat_f[i, j] = r
        
        # ç”·æ€§
        v1m, v2m = male_vectors[y1], male_vectors[y2]
        valid_m = ~(np.isnan(v1m) | np.isnan(v2m))
        if np.sum(valid_m) >= 10:
            r_m, _ = pearsonr(v1m[valid_m], v2m[valid_m])
            corr_mat_m[i, j] = r_m

# ======================
# ç»˜åˆ¶çƒ­åŠ›å›¾
# ======================
plt.rcParams['font.sans-serif'] = ['SimHei']  # æ”¯æŒä¸­æ–‡
plt.rcParams['axes.unicode_minus'] = False   # æ­£å¸¸æ˜¾ç¤ºè´Ÿå·

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# å¥³æ€§çƒ­åŠ›å›¾
sns.heatmap(
    corr_mat_f,
    xticklabels=[str(y) for y in years_list],
    yticklabels=[str(y) for y in years_list],
    cmap="RdBu_r",
    center=0.5,
    vmin=0.0,
    vmax=1.0,
    square=True,
    cbar_kws={'label': 'Pearson ç›¸å…³ç³»æ•°'},
    ax=axes[0]
)
axes[0].set_title("å¥³æ€§ç¾¤ä½“è¡¨å¾çš„å¹´åº¦ç¨³å®šæ€§ (1950â€“2019)", fontsize=14)
axes[0].set_xlabel("å¹´ä»½")
axes[0].set_ylabel("å¹´ä»½")
axes[0].tick_params(axis='x', rotation=90, labelsize=8)
axes[0].tick_params(axis='y', rotation=0, labelsize=8)

# ç”·æ€§çƒ­åŠ›å›¾
sns.heatmap(
    corr_mat_m,
    xticklabels=[str(y) for y in years_list],
    yticklabels=[str(y) for y in years_list],
    cmap="RdBu_r",
    center=0.5,
    vmin=0.0,
    vmax=1.0,
    square=True,
    cbar_kws={'label': 'Pearson ç›¸å…³ç³»æ•°'},
    ax=axes[1]
)
axes[1].set_title("ç”·æ€§ç¾¤ä½“è¡¨å¾çš„å¹´åº¦ç¨³å®šæ€§ (1950â€“2019)", fontsize=14)
axes[1].set_xlabel("å¹´ä»½")
axes[1].set_ylabel("å¹´ä»½")
axes[1].tick_params(axis='x', rotation=90, labelsize=8)
axes[1].tick_params(axis='y', rotation=0, labelsize=8)

plt.tight_layout()
output_path = "analysis3_temporal_stability_heatmap.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')
plt.show()

print(f"âœ… çƒ­åŠ›å›¾å·²ä¿å­˜è‡³: {os.path.abspath(output_path)}")
```
